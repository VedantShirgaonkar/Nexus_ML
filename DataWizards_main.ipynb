```python
df1 = df.drop(['TargetID','DRUGID','PUBCHCID','Accession Number','UNIPROID','SYNONYMS','FUNCTION','DRUGTYPE','TARGNAME'],axis=1)
```

```python
df1.shape
```

```python
from sklearn.preprocessing import LabelEncoder

le1 = LabelEncoder()
for column in df1.columns:
    if df1[column].dtype == 'object':
        df1[column] = le1.fit_transform(df1[column])
```

```python
test1 = test.drop(['TargetID','DRUGID','PUBCHCID','Accession Number','UNIPROID','SYNONYMS','FUNCTION','DRUGTYPE','TARGNAME'],axis=1)
```

```python
test1.shape
```

```python
from sklearn.preprocessing import LabelEncoder

le2 = LabelEncoder()
for column in test1.columns:
    if test1[column].dtype == 'object':
        test1[column] = le2.fit_transform(test1[column])
```

# 1. Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier

X_train = df1.drop(['ID','Target_Status'], axis=1)
y_train = df1['Target_Status']

model_1 = RandomForestClassifier(random_state=42)
model_1.fit(X_train, y_train)
```

```python
X_test = test1.drop(['ID'],axis=1)
y_pred_1 = model_1.predict(X_test)
```

```python
Y_series_1 = pd.Series(le1.inverse_transform(y_pred_1))
Y_series_1.index = X_test.index
test1['Prediction'] = pd.Series(Y_series_1,index = Y_series_1.index)
true_df_1 = test1[['ID','Prediction']].iloc[Y_series_1.index]
```

```python
true_df_1.to_csv('Submission_1.csv',index=False)
```

# 2. XGB Classifier

```python
from xgboost import XGBClassifier

model_2 = XGBClassifier(random_state=42)
model_2.fit(X_train, y_train)
```

```python
y_pred_2 = model_2.predict(X_test)
```

```python
Y_series_2= pd.Series(le1.inverse_transform(y_pred_2))
Y_series_2.index = X_test.index
test1['Prediction'] = pd.Series(Y_series_2,index = Y_series_2.index)
true_df_2 = test1[['ID','Prediction']].iloc[Y_series_2.index]
```

```python
true_df_2.to_csv('Submission_2.csv',index=False)
```
