```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import f1_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.naive_bayes import GaussianNB
```

```python
train_df = pd.read_csv('/Users/rahul/Desktop/Nexus_Datasets/train.csv')
test_df = pd.read_csv('/Users/rahul/Desktop/Nexus_Datasets/test.csv')
```

```python
train1_df = train_df.copy()
train1_df = train1_df.drop(['TargetID','DRUGID','PUBCHCID','DRUGTYPE','FUNCTION','Accession Number','SYNONYMS','UNIPROID'], axis=1)
train1_df.head(5)
```

```python
# Label encoding all object (categorical) columns
label_encoders = {}
for col in train1_df.columns:
    le = LabelEncoder()
    train1_df[col] = le.fit_transform(train1_df[col])
    label_encoders[col] = le  # Save the encoder for future use (like transforming test data)
```

```python
# Separate features (X) and target (y)
X = train1_df.drop(['Target_Status','ID'], axis=1)
y = train1_df['Target_Status']
```

```python
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2)
```

```python
# Initialize and train Random Forest
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)
```

```python
# Predict and calculate F1 Score
y_pred = rf.predict(X_val)
f1_rf = f1_score(y_val, y_pred, average='weighted')
print(f"F1 Score (Random Forest): {f1_rf}")
```
